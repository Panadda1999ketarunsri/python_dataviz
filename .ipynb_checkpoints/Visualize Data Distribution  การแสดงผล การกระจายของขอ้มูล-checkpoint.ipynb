#  Scatter plot

```python
irisdata = pandas.read_csv('C:\\Users\\acer\\Desktop\\Basic programing\\iris.data',header = None)
irisdata                                                                      
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

```python
# ตารางแบบย่อๆ
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])  #[แถว คือ แนวนอน ,]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6)  # sคือ ขนาดของจุด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=80) # ปรับขนาด s ให้ใหญ่ขึ้น
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])  # s=10*setosa_np[:,2] อันนี้คือทำให้เห็นชัดขึ้นว่า แต่ละอันใหญ่ไม่เท่ากัน
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'c')  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'y')  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'g')  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'r')  
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i< 0.15:
        colors.append('b')
    elif i< 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = colors)  # การทำให้จุด มีสีได้หลายสี
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3])  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap = 'YlOrBr')  # 'YlOrBr' ค่า น้อย ไป มาก 
                                                                                                 #Y เหลืองคือ ค่าข้อมูลน้อยสุด
```

# การใช้ค่า RGB ในการแสดงผล 

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]])) # การผสมสี สามสี แดงเขียวฟ้า โดยกำหนดสัดส่วนของความเข้มสีได้
                                                                                 # c = np.array([[1,0.5,0.8]เช่นให้สีแรกเข้มสุด สีฟ้าเท่ากับ 1
```

# markerstyle รูปแบบของจุด 

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap = 'YlOrBr',marker='*') #marker คือ ลักษณะของจุด
```
