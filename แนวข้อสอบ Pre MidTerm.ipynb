สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
- ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }

#### สร้าง function ชื่อ createDict 
  โดยเราจะใช้ def ในการสร้าง  
  function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
          output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
                 def createDict(listtext, listnumber, logicV)

```python
def createDict(listtext, listnumber, logicV):    # ( [ 'apple' , 'banana'] , [1,2] , True)คือเวลาคีย์ ให้คีย์สามอันแบบนี้ #function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
    if  logicV:  #logicV == True                 # เป็นการลดรูปให้เขียนสั้นขึ้น # Key ที่เอาไว้ชี้ index 
        retdict = {}
        for index in range(len(listtext)):      
            retdict[listtext[index]] = listnumber[index]
    else:    
        retdict = {}
        for index in range(len(listtext)):      
             retdict[listnumber[index]]   = listtext[index]  # false คือเอา retdict[listtext[index]] = listnumber[index] มาสลับกัน
    
    return retdict                                           #ส่งค่า output
```

```python
createDict( [ 'apple' , 'banana'] , [1,2] , True)   #ทดสอบดูว่าใช้ได้ไหม 
```

```python
createDict( [ 'apple' , 'banana'] , [1,2] , False)
```

###  ให้บอกว่ากรดอะมีโนแต่ละอันมีกี่ตัว เช่น A มีกี่ตัว

```python
def countChar(seq,character):   # สร้างฟังก์ชัน ขึ้นว่า เพื่อนับตัวอักษร
    charcount = 0
    for char in seq:
        if char == character:
            charcount =  charcount + 1
    return  charcount
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGATCTCTTGTAGATCTGTTCTCTAAA','A')   # ทดสอบนับดูว่า มี A ทั้งหมดกี่ตัว
```

```python
ANum= 0
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]):                 #วนลูปเพื่อนับแต่ละแถว shape คือ แถว
    ANum += countChar(DNAseq.iloc[row][0],'A')    # คำสั่งย่อ DNAlong = DNAlong += len(DNAseq.iloc[row][0])
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
 
print(f'There are {ANum} As,{TNum} Ts, {GNum} Gs, {CNum} Cs')
```

### ลำดับที่ติดกัน 5 ตัวที่ซ้ำกันมากที่สุด

```python
'ATTAAAGGTTT'  # 5 ตัวติดกัน 'ATTAAAGGTTT' 
```

```python
def populate5charstring(seq):                    # ลูปแรกคือวนว่า 5 ตัวที่ติดกันมีอะไรบ้าง
    str_list = list()
    for i in range(len(seq)-4):
        str_list.append(seq[i:i+5])
    return str_list
    
```

```python
populate5charstring('ATTAAAGGTTT')
```

```python
# 'ATTAAAGGTTT' = ['ATTAA', 'TTAAA', 'TAAAG', 'AAAGG', 'AAGGT', 'AGGTT'] คือนับ 5 ตัวติดแล้วเลื่อนไปทีละตัว
```

```python
list_all_5char = list()
for row in range(DNAseq.shape[0]):
    list_all_5char =   list_all_5char + populate5charstring(DNAseq.iloc[row][0])
```

```python
len(list_all_5char)
```

```python
unique_5char = list(set(list_all_5char ))           
```

```python
len(unique_5char)
```

```python
dict_5char = {}
for sub_str in unique_5char:       # จะนับว่าแต่ละอันมีซ้ำกันเท่าไหร่
    num = 0
    for s in list_all_5char:
        if sub_str == s:
            num += 1
    dict_5char[sub_str] = num
```
